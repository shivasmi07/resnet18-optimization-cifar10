```python
import pandas as pd
import matplotlib.pyplot as plt

grad = pd.read_csv('GD_fitness.csv')
pso = pd.read_csv('PSO_fitness.csv')
cma = pd.read_csv('CMA_fitness.csv')

    
grad = grad.columns.to_numpy().astype(float)
pso = pso.iloc[:,0]
cma = cma.columns.to_numpy().astype(float)


plt.plot(grad, 'r', label = 'Gradient Descent')
plt.plot(pso, 'b', label = 'PSO')
plt.plot(cma, 'g', label = 'CMA-ES')
plt.title('Change in Loss of Network over Generations')
plt.xlabel('Generations')
plt.ylabel('Loss')
plt.legend()
plt.figure(figsize =(2000,2000))
plt.show()
```
